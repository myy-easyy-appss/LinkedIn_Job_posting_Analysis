```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce
from wordcloud import WordCloud
```

##### Data Collection and Pre-Processing

```python
job_postings = pd.read_csv("job_postings.csv")
job_postings
```

##### let's check if there is duplicate and null in Job_id

```python
duplicates = job_postings['job_id'].duplicated()
num_duplicates = duplicates.sum()
num_duplicates
```

```python
job_postings['job_id'].isnull().sum()
```

##### As there are no duplicates and nulls so the next step is to merge files first is job_benefits file

```python
job_benefits = pd.read_csv("benefits.csv")
job_benefits = job_benefits.drop('inferred',axis = 1)
job_benefits = job_benefits.groupby('job_id')['type'].agg(lambda x: ' , '.join(x)).reset_index()
```

```python
job_postings = job_postings.merge(job_benefits, on = "job_id", how = "left")
job_postings
```

```python
job_skills = pd.read_csv("job_skills.csv")
job_skills = job_skills.groupby('job_id')['skill_abr'].agg(lambda x: ' , '.join(x)).reset_index()
job_postings = job_postings.merge(job_skills, on="job_id", how = "left")
job_postings
```

##### Pre-processing the company details

```python
company_data = pd.read_csv("companies.csv")

company_industries_d = pd.read_csv("company_industries.csv")
company_industries_d = company_industries_d.groupby('company_id')['industry'].agg(lambda x: ', '.join(x)).reset_index() 

company_specialities_d = pd.read_csv("company_specialities.csv")
company_specialities_d = company_specialities_d.groupby('company_id')['speciality'].agg(lambda x: ', '.join(x)).reset_index() 
employee_counts_d = pd.read_csv("employee_counts.csv")
employee_counts_d = employee_counts_d.groupby('company_id')['time_recorded'].max().reset_index() 

company_data = company_data.merge(company_industries_d, on="company_id", how="left")
company_data = company_data.merge(company_specialities_d, on="company_id", how="left")
company_data = company_data.merge(employee_counts_d, on="company_id", how="left")
company_data
```

##### Merging of job_postings and Company_data

```python
merged = job_postings.merge(company_data, on = 'company_id', how = "left")
merged
```

```python
column_names = merged.columns
print(column_names)
```

```python
choosed_columns = ['title','name','description_x','formatted_work_type','location','original_listed_time','application_type','sponsored','description_y','company_size','industry','skill_abr','state', 'country', 'max_salary',
       'med_salary', 'min_salary', 'formatted_experience_level']
merged = merged[choosed_columns]
merged
```

```python
pretty_column_name = {'title': 'job_title', 'name': 'company_name', 'description_x': 'job_description',
               'formatted_work_type': 'work_type','original_listed_time': 'listed_time','description_y': 'company_description'}
data = merged.rename(columns=pretty_column_name)

data
```

# EDA

```python
# Create a word cloud from job titles
job_titles_text = ' '.join(data['job_title'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(job_titles_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Job Title Word Cloud')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Company Size Distribution (Pie Chart)
company_size_counts = data['company_size'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(company_size_counts, labels=company_size_counts.index, autopct='%1.1f%%', colors=['coral','lightgreen','blue','purple','red', 'yellow','olive'])
plt.title('Company Size Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Job Type Breakdown (Bar Chart)
job_type_counts = data['work_type'].value_counts()
plt.figure(figsize=(10, 6))
job_type_counts.plot(kind='bar', color='lightseagreen')
plt.title('Job Type Breakdown')
plt.xlabel('Job Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Visualization 6: Sponsored vs. Non-Sponsored Listings (Pie Chart)
sponsored_counts = data['sponsored'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sponsored_counts, labels=['Non-Sponsored', 'Sponsored'], autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Sponsored vs. Non-Sponsored Listings')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Top 6 Industries with Highest Job Demand
import seaborn as sns
top_industries = data['industry'].value_counts().head(6)

# Convert to DataFrame for heatmap
industry_df = pd.DataFrame({'industry': top_industries.index, 'Job Listings': top_industries.values})

# Reshape the data for a heatmap
industry_df = industry_df.pivot_table(values="Job Listings", index="industry")

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(industry_df, annot=True, cmap="Blues", fmt=".0f", linewidths=0.5)

# Title
plt.title("Top 6 Industries with Highest Job Demand")

# Show the plot
plt.show()
```

```python
# Count job postings per industry and job title
industry_jobs = data.groupby(['industry', 'job_title']).size().reset_index(name='total_postings')

# Sort within each industry to find the most common jobs
industry_jobs = industry_jobs.sort_values(by=['industry', 'total_postings'], ascending=[True, False])

# Display the top results
print(industry_jobs.head(20))  # Adjust to see more results

top_industries = industry_jobs.groupby('industry')['total_postings'].sum().nlargest(10).index
top_jobs = industry_jobs[industry_jobs['industry'].isin(top_industries)]

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_jobs, x='total_postings', y='job_title', hue='industry', dodge=False)
plt.xlabel("Number of Postings")
plt.ylabel("Job Titles")
plt.title("Most Common Job Titles in Each Industry")
plt.legend(title="Industry")
plt.show()
```
